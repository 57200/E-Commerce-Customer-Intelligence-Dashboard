```python
date_columns = [
    "review_creation_date",
    "review_answer_timestamp"
]

for col in date_columns:
    reviews[col] = pd.to_datetime(reviews[col])
```

```python
reviews[reviews.duplicated(subset=["review_id"], keep=False)].sort_values("review_id")
```

## Duplicate Analysis

### Observations

- No fully duplicated rows were found.
- `review_id` appears duplicated in 814 records.
- Investigation showed that the same `review_id` can be associated with different `order_id` values while retaining the same review content.

### Decision

- These are not data quality issues.
- Duplicate `review_id` values represent one review linked to multiple orders.
- Therefore, no records were removed.
