# E-Commerce Customer Intelligence Dashboard

## Notebook 06 – Data Integration

### Objective

- Understand relationships between datasets
- Identify primary and foreign keys
- Merge cleaned datasets
- Create a Master Analytical Dataset

```python
import pandas as pd

orders = pd.read_csv("../data/cleaned/orders_cleaned.csv")
customers = pd.read_csv("../data/cleaned/customers_cleaned.csv")
payments = pd.read_csv("../data/cleaned/payments_cleaned.csv")
products = pd.read_csv("../data/cleaned/products_cleaned.csv")
reviews = pd.read_csv("../data/cleaned/reviews_cleaned.csv")

# Order Items (Raw dataset)
order_items = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
```

```python
orders.shape
customers.shape
payments.shape
products.shape
reviews.shape
order_items.shape
```

```python
order_items.info()
```

```python
order_items.isnull().sum()
```

```python
order_items.duplicated().sum()
```

```python
order_items.duplicated(
    subset=["order_id","order_item_id"]
).sum()
```

```python
master = pd.merge(
    orders,
    customers,
    on="customer_id",
    how="left"
)
print(master.shape)
```

```python
master = pd.merge(
    master,
    payments,
    on="order_id",
    how="left"
)
print(master.shape)
```

```python
master = pd.merge(
    master,
    reviews,
    on="order_id",
    how="left"
)
print(master.shape)
```

```python
master = pd.merge(
    master,
    order_items,
    on="order_id",
    how="left"
)
print(master.shape)
```

```python
master = pd.merge(master, products, on="product_id", how="left")
print(master.shape)
```

```python
master.head()
```

```python
master.info()
```

```python
master.shape
```

```python
master.to_csv(
    "../data/cleaned/master_dataset.csv",
    index=False
)
```
