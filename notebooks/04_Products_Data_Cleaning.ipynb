## Missing Value Analysis

### Observations

- 610 products have missing product category and related descriptive information.
- All 610 products appear in the `order_items` dataset, indicating that they were purchased by customers.
- Two products have missing physical dimensions (weight/length/height/width).

### Decision

- Retain all 610 products because removing them would result in the loss of valid sales records.
- Retain the two products with missing dimensions since these attributes are not critical for customer behavior analysis.
- Missing category values will be handled during visualization by labeling them as **"Unknown Category"** where appropriate.
